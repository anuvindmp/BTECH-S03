```python
def first2last2(str1):
    if len(str1) < 2:
        return ""
    return str1[:2] + str1[-2:]
```

```python
str1 = input("Enter a string : ")
first2last2(str1)
```

```python
#a
def dollar(str1):
    return str1[0] + str1.replace(str1[0],"$")[1:]

str1 = input("Enter a string : ")
dollar(str1)
```

```python
#b
def removeAtIndex(str1, index):
    if len(str1) < 1:
        return "empty string !!"
    elif len(str1) <= index or index < 0:
        return "invalid index"
    str1 = str1[:index] + str1[index+1:]
    return str1


str1 = input("enter a string")
index = int(input("enter the index of the element to be removed : "))
removeAtIndex(str1, index)
```

```python
#c
def exchange_First_Last(str1):
    if len(str1) < 1:
        return "empty string !!"
    return str1[-1] + str1[1:len(str1)-1] + str1[0]

str1 = input("Enter a string : ")
exchange_First_Last(str1)
```

```python
#d
s = "hi there hello there hi you my yes you"
words = s.split()
wordcount=dict()
for i in words:
    if i not in wordcount:
        wordcount[i]=1
    else:
        wordcount[i]+=1

for i in wordcount:
    print("{:<8}:{:>4}".format(i,wordcount[i]))
```

```python
#a
def Replace_vowels(word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    for i in vowels:
        word = word.replace(i,"")
    return word

Replace_vowels("orange")
```

```python
#b
def modify_case(word):
    word = list(word)
    for i in range(len(word)):
        if word[i].isupper():
            word[i] = word[i].lower()
        else:
            word[i] = word[i].upper()
    return ''.join(word)

print(modify_case("aBUaOP"))  
```

```python
def palindromeCheck(word):
    if word.lower() == word.lower()[::-1]:
        print(f"{word} is a palindrome")
    else:
        print(f"{word} is not a palindrome")

word = input("Enter a string : ")
palindromeCheck(word)
palindromeCheck("dad")
palindromeCheck("madam")
palindromeCheck("python")
```

```python
def print_initials(name):
    initials = name[0].upper()+"."
    for i in range (len(name)):
        if name[i] == " ":
            initials += name[i+1].upper()+"."
    return initials

name = input("Enter your full name: ")
print_initials(name)
```

```python
def passwordCheck(word):
    flag = [0, 0, 0]
    if len(word) < 8 :
        print("password must be atleast 8 characters long")
        return False
    for i in range(len(word)):
        if word[i].isdigit():
            flag[0]=1
        if word[i].isupper():
            flag[1]=1
        if word[i].islower():
            flag[2]=1
    if flag==[1, 1, 1]:
        print("Safe password")
        return True
    else:
        print("password must contain atleast one uppercase,lowercase and numeric character")
        return False
    

password = input("Enter new password :")
passwordCheck(password)
```
