```python
def first2last2(str1):
    if len(str1) < 2:
        return ""
    return str1[:2] + str1[-2:]
```

```python
str1 = input("Enter a string : ")
first2last2(str1)
```

```python
#a
def dollar(str1):
    return str1[0] + str1.replace(str1[0],"$")[1:]

str1 = input("Enter a string : ")
dollar(str1)
```

```python
#b
def removeAtIndex(str1, index):
    if len(str1) < 1:
        return "empty string !!"
    elif len(str1) <= index or index < 0:
        return "invalid index"
    str1 = str1[:index] + str1[index+1:]
    return str1


str1 = input("enter a string")
index = int(input("enter the index of the element to be removed : "))
removeAtIndex(str1, index)
```

```python
#c
def exchange_First_Last(str1):
    if len(str1) < 1:
        return "empty string !!"
    return str1[-1] + str1[1:len(str1)-1] + str1[0]

str1 = input("Enter a string : ")
exchange_First_Last(str1)
```

```python
#d
s = "hi there hello there hi you my yes you"
words = s.split()
wordcount=dict()
for i in words:
    if i not in wordcount:
        wordcount[i]=1
    else:
        wordcount[i]+=1

for i in wordcount:
    print("{:<8}:{:>4}".format(i,wordcount[i]))
```
