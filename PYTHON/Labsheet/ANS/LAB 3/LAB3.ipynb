```python
def positionSwap(lst, p1, p2):
    lst[p1],lst[p2] = lst[p2],lst[p1]
    return lst

nums = []
n = int(input("Enter the length of your list : "))
for i in range(n):
    num = int(input("Enter a number: "))
    nums.append(num)
print("Original list : ", nums)

pos1 = int(input("enter the index to be swapped"))
pos2 = int(input("enter the index to be swapped"))
print(f"List after swapping {pos1} and {pos2} : ", positionSwap(nums,pos1,pos2))
```

```python
def exchange(lst):
    lst[:len(lst)//2],lst[len(lst)//2:] = lst[len(lst)//2:],lst[:len(lst)//2] 
    return lst

nums = [1,2,3,4]
exchange(nums)
```

```python
def peakElement(nums):
    peak = []
    if nums[0] > nums[1]:
        peak.append(nums[0])
    for i in range(1, len(nums) - 1):
        if nums[i] > nums[i - 1] and nums[i] > nums[i + 1]:
            peak.append(nums[i])   
    if nums[-1] > nums[-2]:
        peak.append(nums[-1])

    return peak

nums = [15, 30, 20, 5]
print("peak element is/are :",peakElement(nums))
```

```python
def duplicateCheck(nums):
    dup = []
    notDup = []
    for i in range(len(nums)-1):
        if nums.count(nums[i]) > 1 :
            dup.append(nums[i])
        else:
            notDup.append(nums[i])
    dup = list(set(dup))
    print("Duplicate elements are :",dup)
    print("Elements withot duplicates are:",notDup)

nums = [2,4,3,2,1,6,7,6,2,2,4,3]
duplicateCheck(nums)
```

```python
def rotate(nums):
    numsCopy = nums.copy()
    nums[-1] = numsCopy[0]
    for i in range (len(numsCopy)-1):
        nums[i] = numsCopy[i+1]
    return nums

nums = [1,2,3,4]
rotate(nums)
```

```python
def sumPair(l1,l2,x):
    sum = []
    for i in l1:
        for j in l2:
            if j + i == x:
                sum.append((i, j))
    return sum

list1 = [-1,-2,4,-6,5,7]
list2 = [6,3,4,0]
x = 8

sumPair(list1,list2,x)
    
```

```python
def read(rows, cols):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = int(input(f"Enter element for row {i + 1}, column {j + 1}:"))
            row.append(element)
        matrix.append(row)
    return matrix

def display(matrix):
    for row in matrix:
        print(" ".join(str(element) for element in row))

def sumRow(matrix):
    rowSums = [sum(row) for row in matrix]
    return rowSums

def sumCol(matrix):
    col_sums = [sum(column) for column in zip(*matrix)]
    return col_sums

def add(matrix1, matrix2):
    result_matrix = [[matrix1[i][j] + matrix2[i][j] for j in range(len(matrix1[0]))] for i in range(len(matrix1))]
    return result_matrix

rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

print("Enter Matrix 1:")
matrix1 = read(rows, cols)
print("Enter Matrix 2:")
matrix2 = read(rows, cols)
print("\nMatrix 1:")
display(matrix1)
print("\nMatrix 2:")
display(matrix2)


print("\nSum of elements in each row of Matrix 1:", sumRow(matrix1))
print("Sum of elements in each column of Matrix 1:", sumCol(matrix1))
print("\nSum of elements in each row of Matrix 2:", sumRow(matrix2))
print("Sum of elements in each column of Matrix 2:", sumCol(matrix2))
print("\nSum of Matrix 1 and Matrix 2:")
matrixSum = add(matrix1, matrix2)
display(matrixSum)
```
